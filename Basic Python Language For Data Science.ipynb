While Loop

```python
#while loop— repeating code while a condition is True
#while loop digunakan untuk menjalankan kode berulang kali selama kondisi bernilai True.
i = 0 #inilization
while i<5: #condition
  print(i)
  i+=1 #increment
```

Penjelasan:

Initialization: i = 0 → memulai variabel counter

Condition: i < 5 → loop berjalan selama True

Body: print(i) → perintah yang dijalankan setiap iterasi

Increment: i += 1 → agar kondisi akhirnya menjadi False dan loop berhenti

```python
#loops and control statement
#break -- this statement break the loop
#break digunakan untuk menghentikan loop secara langsung, walaupun kondisi loop masih True.
i = 0
while i<5:
  i+=1 #increment
  if i==3:
    break
  else:
    print(i)
```

Penjelasan:

i = 0 → mulai

Loop while i < 5 berjalan

i += 1 → menambah i setiap iterasi

if i == 3: → saat i = 3, break menghentikan loop

else: print(i) → hanya mencetak i sebelum i = 3

```python
#continue statement-- this statement skip the current iteration
#continue digunakan untuk melewatkan sisa kode di dalam loop saat ini dan melanjutkan ke iterasi berikutnya.
i = 0
while i<5:
  i+=1
  if i==3:
    continue
    print(i)
  else:
    print(i)
```

Penjelasan:

i = 0 → mulai

Loop while i < 5 berjalan

i += 1 → menambah i setiap iterasi

if i == 3: → saat i = 3, continue dilewati, sehingga print(i) di bawah continue tidak dijalankan

Loop melanjutkan ke iterasi berikutnya

```python
#pass statement -- this statement pass the code/this statement does nothing
#pass digunakan sebagai placeholder di Python. Artinya, Python tidak melakukan apa-apa saat mencapai pass.
#Biasanya dipakai untuk struktur kode yang belum diisi agar tidak menimbulkan error.
i = 0
while i<5:
  i+=1
  if i==3:
    pass
    print(i)
  else:
    print(i)
```

Penjelasan:

pass tidak mengubah alur program.

Dalam contoh di atas, loop tetap berjalan normal.

Berguna untuk:

Menulis skeleton kode sementara

Membuat blok if, loop, function tanpa isi untuk sementara

```python
#positional argument-passing values based on position
#Positional arguments adalah nilai yang dikirim ke fungsi berdasarkan urutan parameternya.
def calculation(a, b):
  print("addition", a+b)
  print("subtraction", a-b)
  print("multiplication", a*b)
  print("Division", a/b)
  return "Calculation is complete"
```

```python
print(calculation(17, 5))
```

```python
#default argument— parameters with default values
#Default arguments adalah parameter fungsi yang memiliki nilai bawaan,
#sehingga tidak wajib diberikan saat memanggil fungsi.
def area_perimeter(panjang=4, lebar=8):
  luas = panjang*lebar
  perimeter = 2*(panjang + lebar)
  return "Area is", luas, "and perimeter is", perimeter
```

```python
print(area_perimeter())
```

```python
print(area_perimeter(10, 20))
```

```python
print(area_perimeter(10,10))
```

```python
#keyword argument—passing values using parameter names
#Keyword arguments adalah cara memanggil fungsi dengan menyebutkan nama parameternya,
#sehingga urutan tidak lagi penting.
def interest(p, r, t):
  i = (p*r*t)/100
  return i
print(interest(r = 10, t = 2, p = 1000))
```

```python
#mixed argument (campuran positional + keyword)
#positional argument follows keyword argument adalah error yang muncul ketika kamu menulis argumen posisi setelah argumen keyword, yang tidak diperbolehkan.
print(interest(1000, t = 2, r = 10))
```

```python
print(interest(r = 10,p=1000,  2))
```

```python
def calculation(a, b):
    print("Addition:", a + b)
    print("Subtraction:", a - b)
    print("Multiplication:", a * b)
    print("Division:", a / b)
calculation(a=10, b=5)
```

```python
#variable length argument-argumen yang jumlahnya tidak tetap.
#non keyword positional variable length argument
def test(*args):
  print(args)
  print(len(args))
  print(type(args))
```

```python
test(2,3, 4, 8, 9, 7)
```

```python
def sum_number(*args):
  sum = 0
  for num in args:
    sum+=num
  return sum
```

```python
sum_number(3, 5, 6, 7, 8, 9)
```

```python
#keyword variable length argument(kwargs)
#**kwargs memungkinkan kita menerima argumen dengan keyword dalam jumlah tidak terbatas.
def test(**kwargs):
  print(kwargs)
  print(type(kwargs))
  print(len(kwargs))
```

```python
test(a = 10, b = 20, c = 30)
```

```python
import matplotlib.pyplot as plt
import numpy as np
fruit = np.array(["Apple", "Banana", "Orange", "Watermelon"])
quantity = np.array([100, 120, 80, 50])
color = np.array(["red", "yellow", "orange", "green"])
plt.bar(fruit, quantity, color = color)
plt.xlabel("Fruit")
plt.ylabel("Quantity")
plt.title("Fruit Quantity")
plt.show()
```

```python
fruit = np.array(["Apple", "Banana", "Orange", "Watermelon"])
quantity = np.array([100, 120, 80, 50])
color = np.array(["red", "yellow", "orange", "green"])
plt.barh(fruit, quantity, color = color)
plt.xlabel("Quantity")
plt.ylabel("Fruit")
plt.title("Fruit Quantity")
plt.show()
```

```python
#pie plot
quantity = np.array([20, 35, 15, 30])
fruit = np.array(["Apple", "Banana", "Orange", "Mango"])
myexplode = [0.2, 0, 0, 0]
plt.pie(quantity, labels = fruit, explode = myexplode)
plt.legend(title = "Fruit",loc = "best")
plt.show()
```

```python
#scatter plot
age_of_car = np.array([4, 6, 8, 10, 12, 14, 16, 18])
speed_of_car = np.array([150, 130, 120, 110, 100, 90, 80, 70])
plt.scatter(age_of_car, speed_of_car)
plt.xlabel("Age of car")
plt.ylabel("Speed of car")
plt.title("Age and speed of car")
plt.show()
```

```python
#histogram-- it shows frequency distribution
import numpy as np
x = np.random.randint(40, 100, (1, 200))
print(x)
x = x.flatten()
print(x)
```

```python
plt.hist(x, bins=10, color="skyblue", edgecolor="black")
plt.title("Histogram Nilai")
plt.xlabel("Nilai")
plt.ylabel("Frekuensi")
plt.show()
```
